## Precipitation by month 2007 - 2018 

```python
# Read precipitation csv file to dataframe
wclim_prec = pd.read_csv(path+'precipitation2007_2018.csv',encoding="latin1")
# Remove extra column
wclim_prec = wclim_prec.drop('Unnamed: 0', 1)
```

```python
# Preview
wclim_prec.head()
```

Reordering columns chronologically

```python
cols=wclim_prec.columns.tolist()
#len(cols)
cols
```

```python
## each year has 12 months -> for loop test for month and year
months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
years =['07','08','09','10','11','12','13','14','15','16','17','18']
columns_order_prec=[]
for year in years:
  for i in range (0,12):    ##0-11
     columns_order_prec.append('PRECIPITATION_' + months[i] + '_' + year)
columns_order_prec.extend( ['COD_DPTO',
 'NOM_DPTO',
 'COD_MPIO',
 'NOM_MPIO',
 'TIPO',
 'LATITUD',
 'LONGITUD'])
columns_order_prec
```

```python
#re-assign ordered columns to dataframe
wclim_prec = wclim_prec[columns_order_prec]
wclim_prec
```

## Temperature by month 2007 - 2018 

```python
# Read temperature csv file to dataframe  
wlcim_temp = pd.read_csv(path+'tempearture_2007_2018.csv',encoding="latin1")
wlcim_temp = wlcim_temp.drop('Unnamed: 0', 1)
```

```python
# Preview
wlcim_temp.head()
```

Reordering Columns Chronologically

```python
## each year has 12 months -> for loop test for month and year
months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
years =['07','08','09','10','11','12','13','14','15','16','17','18']
columns_order_temp=[]
for year in years:
  for i in range (0,12):    ##0-11
      columns_order_temp.append('TEMPERATURE_' + months[i] + '_' + year)
columns_order_temp.extend( ['COD_DPTO',
 'NOM_DPTO',
 'COD_MPIO',
 'NOM_MPIO',
 'TIPO',
 'LATITUD',
 'LONGITUD'])
columns_order_temp
```

```python
#re-assign ordered columns to dataframe
wlcim_temp = wlcim_temp[columns_order_temp]
wlcim_temp
```

## Elevation

```python
# Read elevation csv file to dataframe  
wlcim_elevation = pd.read_csv(path+'elevation.csv', encoding="latin1")
wlcim_elevation = wlcim_elevation.drop('Unnamed: 0', 1)
```

```python
# Preview
wlcim_elevation.head()
```

# Join Worldclim temperature + precipitation + elevation

```python
# merge temperature + precipitation using 'COD_DPTO',	'NOM_DPTO',	'COD_MPIO',	'NOM_MPIO',	'TIPO',	'LATITUD',	'LONGITUD'
worldclim_tem_prec_df = wclim_prec.merge(wlcim_temp, on=['COD_DPTO',	'NOM_DPTO',	'COD_MPIO',	'NOM_MPIO',	'TIPO',	'LATITUD',	'LONGITUD'], how='inner')
# merge temperature + precipitation + elevation 'COD_DPTO',	'NOM_DPTO',	'COD_MPIO',	'NOM_MPIO',	'TIPO',	'LATITUD',	'LONGITUD'
worldclim_tem_prec_elev_df = worldclim_tem_prec_df.merge(wlcim_elevation, on=['COD_DPTO',	'NOM_DPTO',	'COD_MPIO',	'NOM_MPIO',	'TIPO',	'LATITUD', 'LONGITUD'], how='inner')
```

```python
# Preview
worldclim_tem_prec_elev_df.head()
```

# Join WorldClim temperature + precipitation + elevation + weekly Dengue Data

```python
# Rename COD_MPIO to Municipality code to merge using Municipality code
worldclim_tem_prec_elev_df = worldclim_tem_prec_df.rename(columns={'COD_MPIO' : 'Municipality code'})
```

## Dengue Weekly cases

```python
# Dengue data csv file to dataframe 
cases_month_week_df = pd.read_csv(path + 'DANE_Dengue_Data_Variables_v3.csv', encoding="latin1")
```

```python
# Rename Municipalitycode to Municipality code to merge using Municipality code
cases_month_week_df = cases_month_week_df.rename(columns={'Municipalitycode': 'Municipality code'})
cases_month_week_df.head()
```

```python
# Merge
worldclim_tem_prec_cases_month_week_df = worldclim_tem_prec_elev_df.merge(cases_month_week_df, on=['Municipality code'], how='inner')
```

```python
# Preview
worldclim_tem_prec_cases_month_week_df.head()
```

```python
# see final columns
for col in worldclim_tem_prec_cases_month_week_df.columns:
  print(col)
```

```python
worldclim_tem_prec_cases_month_week_df.to_csv(r'temperature_precipitation_elevation_DengueCasesWeek.csv')
```
