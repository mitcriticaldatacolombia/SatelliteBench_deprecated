# Files

```python
files = []

for file in os.listdir():
  if file.endswith('.csv'):
    files.append(file)
print(f'The number of files is: {len(files)}')
print(files)
```

```python
def get_variable_and_code(file):
  var = file.split('_')[0]
  # Remove .csv
  code = file.split('_')[1][:-4]
  return var, code
```

```python
from numpy import column_stack
def read_csv(file):
  df = pd.read_csv(file)
  df['LastDayWeek'] = pd.to_datetime(df['LastDayWeek'])
  df = df.set_index('LastDayWeek')
  var, code = get_variable_and_code(file)
  var_name = var + '_' + code
  df.rename(columns={"value": var_name}, inplace=True)
  return df
```

```python
def merge_dfs(df1, df2):
  df = pd.merge(df1, df2, left_index=True, right_index=True)
  return df
```

# Merge dfs

```python
df = pd.DataFrame()

for i, file in enumerate(files):

  # Read file
  df_aux = read_csv(file)

  if i == 0:
    df = df_aux

  else:
    df = merge_dfs(df, df_aux)

var = get_variable_and_code(files[0])[0]
df.to_csv(f'{var}_all.csv')

df
```
